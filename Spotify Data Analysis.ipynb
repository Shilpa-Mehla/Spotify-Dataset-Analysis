#Data Analysis and Visualizations

```python
df['playlist_genre'].value_counts(normalize=True)
```

```python
df['playlist_genre'].value_counts()
```

```python
df.sort_values('track_popularity',ascending=True).head(10)
```

```python
df.query('track_popularity>90',inplace=False).sort_values('track_popularity',ascending=False).head(10)
```

```python
df1=df.sort_values('track_popularity',ascending=False).head(10)
sns.barplot(y='track_popularity',x='playlist_genre',data=df1).set(title='Top Popular Playlist Genre')
```

```python
df['duration']=df['duration_ms'].apply(lambda x: round(x/1000))
df.drop('duration_ms',inplace=True,axis=1)
```

```python
sns.lineplot(y='duration',x='playlist_genre',data=df).set(title='Each Playlist Genre Duration of Songs(in sec)')
```

```python
sns.barplot(y='track_popularity',x='track_artist',data=df1,color='orange').set(title='Top Popular Track Artist')
```

```python
sns.regplot(y='loudness',x='energy',data=df,color='c').set(title='Loudness vs Energy Correlation')
```

```python
sns.regplot(y='track_popularity',x='acousticness',data=df,color='m').set(title='Acousticness vs Track Popularity Correlation')
```

```python
corr_df=df.corr(method='pearson')
sns.heatmap(corr_df).set(title="Correlation Heatmap Among Variables")
```

```python
years=df['track_album_release_date'].dt.year
sns.displot(years,discrete=True,kind='hist').set(title='Number of songs per year')
```

```python
years=df['track_album_release_date'].dt.year
dff=pd.DataFrame(years,df['duration']).reset_index()
sns.barplot(y='duration',x='track_album_release_date',data=dff).set(title='Duration of songs per year')
```
